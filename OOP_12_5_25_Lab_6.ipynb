```python
import pandas as pd


## read file
df = pd.read_csv("data01.csv")
df.head(10)

## drop any missing values
df = df.dropna()

## median of median house value
x = df["median_house_value"].median
print("Median of House Values: ", x)
print("------------------------------------")

## median total rooms
median_tr = df["total_rooms"].median
print("Median of Total Rooms: ", median_tr)
print("------------------------------------")

## replace total rooms less than 500 with median
print("*** Rows with Median Total Rooms ***")
print("------------------------------------")
for x in df.index:
  if df.loc[x, "total_rooms"] < 500:
    df.loc[x, "total_rooms"] = median_tr
    print(df.loc[[x]])
print("---------------------------------------------")

## print housing_median_age less than 25
print("*** Rows with Housing Median Age Below 25 ***")
print("---------------------------------------------")
for x in df.index:
  if df.loc[x, "housing_median_age"] < 25:
    print(df.loc[[x]])
print("------------------------------------")

import matplotlib.pyplot as plt

df.plot(kind='scatter', x='total_bedrooms', y='median_house_value')
plt.show()
```
