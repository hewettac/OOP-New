```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

%matplotlib inline
df = pd.read_csv('homeprices.csv')
df2 = pd.read_csv('homeprices2.csv')

plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color = 'green', marker = '+')
plt.show()

```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price, color = 'green', marker = '+')
plt.plot(df.area, reg.predict(df[['area']]), color = 'navy')
plt.show()
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df2.area, df2.price, color = 'red', marker = '+')
plt.plot(df2.area, reg.predict(df2[['area']]), color = 'blue')
plt.show()
```

```python
df3 = pd.read_csv('areas.csv')
df3

p = reg.predict(df3)
df3['prices'] = p
df3.to_csv('predictions.csv')
```

```python
df4 = pd.read_csv('predictions.csv')
df4

plt.xlabel = ('area')
plt.ylabel = ('price')
plt.scatter(df4.area, df4.prices, color = 'black', marker = '+')
plt.show()
```

```python
plt.xlabel=('area')
plt.ylabel=('price')
plt.scatter(df4.area, df4.prices, color = 'black', marker = '+')
plt.plot(df4.area, reg.predict(df4[['area']]), color = 'orange')
```
