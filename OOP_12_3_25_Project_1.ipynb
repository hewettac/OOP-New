```python
import pandas as pd
df1 = pd.read_csv("data3.csv")

for x in df1.index:
  if df1.loc[x, "Duration"] > 120:
    df1.loc[x, "Duration"] = 100

for x in df1.index:
  if df1.loc[x, "Calories"] < 100:
    df1.loc[x, "Calories"] = 88

for x in df1.index:
  if df1.loc[x, "Calories"] < 500:
    df1.loc[x, "Calories"] = 100

for x in df1.index:
  average_val = (df1.loc[x, "Pulse"] + df1.loc[x, "Maxpulse"])/2
  df1.loc[x, "Average"] = average_val

df1.head(30)


import matplotlib.pyplot as plt

df1.plot()
plt.show()

df1.plot(kind = 'scatter', x = 'Duration', y = 'Calories')
plt.show()

df1.plot(kind = 'bar')
plt.show()

df1.plot(kind = 'hist', x = 'Average')
plt.show()

df1.plot(kind = 'box')
plt.show()
```
