```python
import pandas as pd
import numpy as np

df = pd.read_csv('dataset-01.csv')
count_total  = len(df)

X_a_OOP = (df['OOP']=='A').sum()
P_a_OOP = X_a_OOP/count_total
print("Probability of A in OOP:", P_a_OOP)

X_a_Java = (df['Java']=='A').sum()
P_a_Java = X_a_Java/count_total
print("Probability of A in Java:", P_a_Java)


countofAgivenB = 0
for index, row in df.iterrows():
    if row['Java']=='B' and row['OOP']=='A':
        countofAgivenB += 1
P_B_given_A = countofAgivenB/count_total
print("Probability of B given A:", P_B_given_A)
```
