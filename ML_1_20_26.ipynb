```python
import numpy as np
import pandas as pd

np.random.seed(0)

df=pd.read_csv('dataset1.csv')
x=df[['A','B']]
y=df['C']

class Layer_Dense:
  def __init__(self, n_inputs, n_neurons):
    self.weights=0.1 * np.random.randn(n_inputs, n_neurons)
    self.biases=np.zeros((1, n_neurons))
  def forward(self, inputs):
    self.output=np.dot(inputs, self.weights) + self.biases

class Activation_ReLu:
  def forward(self, inputs):
    self.output=np.maximum(0,inputs)

layer1 = Layer_Dense(2,5)
aa = Activation_ReLu()

layer1.forward(x)
print(layer1.output)
aa.forward(layer1.output)
print(aa.output)

layer2 = Layer_Dense(5,8)
bb = Activation_ReLu()

layer2.forward(aa.output)
print(layer2.output)
bb.forward(layer2.output)
print(bb.output)

layer3 = Layer_Dense(8,2)
cc = Activation_ReLu()

layer3.forward(bb.output)
print(layer3.output)
cc.forward(layer3.output)
print(cc.output)

```
