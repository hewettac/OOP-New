```python
import numpy as np
import pandas as pd

np.random.seed(0)

x=[[2,4],[1,3]]

class Layer_Dense:
  def __init__(self, n_inputs, n_neurons):
    self.weights=0.1 * np.random.randn(n_inputs, n_neurons)
    self.biases=np.zeros((1, n_neurons))
  def forward(self, inputs):
    self.outputU=np.dot(inputs, self.weights) + self.biases

class Activation_ReLu:
  def forward(self, inputs):
    self.outputFU=np.maximum(0,inputs)

layer1 = Layer_Dense(2,3)
aa = Activation_ReLu()

layer1.forward(x)
print("U Value = ", layer1.outputU)
aa.forward(layer1.outputU)
print("F(U)= ",aa.outputFU)

expected_output = 1
Error = pow((aa.outputFU[0][0] - expected_output),2)/2
print("Error is: ", Error)

print("Old Weight for Neuron-One: ", layer1.weights[0][0])
new_weight = layer1.weights[0][0] = layer1.weights[0][0] - (0.5 * Error * x[0][0])
print("New Weight for Neuron-One: ", new_weight)

expected_output = 1
Error = pow((aa.outputFU[0][1] - expected_output),2)/2
print("Error is: ", Error)

print("Old Weight for Neuron-Two: ", layer1.weights[0][0])
new_weight = layer1.weights[0][1] = layer1.weights[0][1] - (0.5 * Error * x[0][0])
print("New Weight for Neuron-Two: ", new_weight)
```
